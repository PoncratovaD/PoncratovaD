```python
import numpy as np
import matplotlib.pyplot as plt

m = np.array([0, 1, 2])
sumsin0 = np.array([-0.3495359632, 0.469612973, 	1.280251271])
sumsin1 = np.array([-0.196825596,	0.6435999359,	1.441472276])
sumsin2 = np.array([-0.1869630841,	0.787505001,	1.574739044])

font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 20,
        }

plt.figure()
plt.plot(m, sumsin0, "--o", label = r'$\theta_{0}$')
plt.plot(m, sumsin1, "--o", label = r'$\theta_{1}$')
plt.plot(m, sumsin2, "--o", label = r'$\theta_{2}$')
plt.xlabel('m', fontdict=font)
plt.ylabel('sin' + r'$\theta$ +  sin' + r'$\theta_{i}$', fontdict=font)
plt.legend()
```
